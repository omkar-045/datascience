```python
import matplotlib.pyplot as plt

virat_years = [2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023, 2024]
odi_runs = [1268,1054,623,739,1460,1202,1377,431,129,302,1377,58]

rohit_years = [2013.0,2014,2015,2016,2017,2018,2019,2020,2021,2022,2023,2024]
runs = [1553,1015,1269,1349,1793,1804,2443,311,1420,995,1800,1154]

rahul_years = [2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023, 2024]
rahul_runs = [0, 4, 252, 196, 52, 69, 572, 443, 177, 251, 1060, 31]

# Create the plot
plt.figure(figsize=(12, 6))  # Optional but recommended for better visuals

plt.plot(virat_years, odi_runs, 'ro--', label="Virat Kohli")     # red circles dashed
plt.plot(rohit_years, runs, 'g^:', label="Rohit Sharma")         # green triangles dotted
plt.plot(rahul_years, rahul_runs, 'bs-.', label="KL Rahul")      # blue squares dash-dot

plt.xlabel("Years")
plt.ylabel("Runs Scored")
plt.title("ODI Performance Comparison (2013–2024)")
plt.legend()
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()

plt.show()
```

```python
plt.style.use('dark_background')
```
