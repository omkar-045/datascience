## Horizontal charsts 

### plt.barh()

```python
import matplotlib.pyplot as plt

# Player names
players = ["Virat Kohli", "Rohit Sharma", "KL Rahul"]

# Runs in the last 5 years (20202024)
kohli_last5 = [431, 129, 302, 1377, 58]              # 2297
rohit_last5 = [311, 1420, 995, 1800, 1154]           # 5680
rahul_last5 = [443, 177, 251, 1060, 31]              # 1962

# Combine all player data in a list of tuples
player_data = [
    ("Virat Kohli", kohli_last5),
    ("Rohit Sharma", rohit_last5),
    ("KL Rahul", rahul_last5)
]

# Extract players and total runs
players = []
runs_5yr = []

# Build the chart data
for name, scores in player_data:
    players.append(name)
    runs_5yr.append(sum(scores))

# Plot horizontal bar chart
plt.figure(figsize=(10, 6))
plt.barh(players, runs_5yr, color=['blue', 'orange', 'green'])

# Add text labels showing equation with .format()
for i in range(len(player_data)):
    name, scores = player_data[i]
    equation = " + ".join(str(x) for x in scores)
    label = "{} = {}".format(equation, sum(scores))
    plt.text(runs_5yr[i] + 50, i, label, va='center', fontweight='bold')

# Chart settings
plt.xlabel("Total Runs (20202024)")
plt.title("ODI Runs in Last 5 Years with Breakdown")
plt.tight_layout()
plt.show()
```
