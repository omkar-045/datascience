# PI charts 

```python
import matplotlib.pyplot as plt 

plt.style.use("ggplot")
```

```python
labels = ["Rohit", 'KL Rahul ', 'kohli ', 'Dhoni']
runs = [18000,8000, 12000,9500 ]

plt.title(" Carieer runs ")
plt.pie(runs , labels = labels )
plt.show()
```

```python
colors = ['#AEDFF7', '#FFDDC1', '#C1E1C1', '#F7C6C7']
plt.pie(runs, 
        labels=labels, 
        colors=colors, 
        wedgeprops={'edgecolor': 'black'})
plt.show()
```

```python
explode=[0.1, 0,0,0]
plt.pie(runs , labels=labels , explode=explode, colors=colors, shadow =True, startangle = 100)
plt.title(" pie examplee")
plt.show()
```

## Crowded Pie Chart

```python
import matplotlib.pyplot as plt

# Example data
labels = ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H']
runs = [40, 15, 10, 7, 5, 4, 3, 2]

# Threshold for grouping small slices
threshold = 5

# Separate large and small slices
large_labels = []
large_runs = []
others_sum = 0

for label, run in zip(labels, runs):
    if run >= threshold:
        large_labels.append(label)
        large_runs.append(run)
    else:
        others_sum += run

if others_sum > 0:
    large_labels.append('Others')
    large_runs.append(others_sum)

plt.pie(large_runs, labels=large_labels, autopct='%1.1f%%', startangle=90)
plt.title('Grouped Pie Chart')
plt.show()
```
